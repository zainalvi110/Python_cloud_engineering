```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('./dataset/titanic.csv')  # Replace with your CSV file name

# Calculate the discounted fare
def calculate_discounted_fare(row):
    fare = row['Fare']
    age = row['Age']
    if age < 20:
        return fare - (fare / 20)  # 5% discount
    elif 20 <= age <= 30:
        return fare - (fare / 30)  # 3.33% discount
    else:
        return fare  # No discount

# Apply the discount calculation to each row
data['Discounted_Fare'] = data.apply(calculate_discounted_fare, axis=1)

data
 
                 
```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv("./dataset/titanic.csv")  # Replace with your CSV file name
def calculate_discounted_fare(row):
    Fare = row['Fare']
    Age = row['Age']
    
    if Age < 20:
        return Fare - (Fare / 20)  # 5% discount
    elif 20 <= Age <= 30:
        return Fare - (Fare / 30)  # 3.33% discount
    else:
        return Fare  # No discount for other ages

# Apply the function to the dataframe
data['Discounted_Fare'] = data.apply(calculate_discounted_fare, axis=1)
data

```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv("./dataset/titanic.csv",index_col=0)  # Replace with your CSV file name
def age_increasing(row):
    Age = row['Age']
    Age =Age+5
    return Age
data['Age'] = data.apply(age_increasing, axis=1)
data 
```
